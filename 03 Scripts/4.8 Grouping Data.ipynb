```python
df_combined['avg_spent'] = df_combined.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
df_combined.loc[df_combined['avg_spent'] >=10, 'Spender_Type'] = 'High spender'
df_combined.loc[df_combined['avg_spent'] < 10, 'Spender_Type'] = 'Low spender'
```

# Step 7

```python
df_combined['median_since_last_order'] = df_combined.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
df_combined.head()
```

```python
df_combined.loc[df_combined['median_since_last_order'] > 20, 'Customer_frequency'] = 'Non-frequent customer'
df_combined.loc[(df_combined['median_since_last_order'] <= 20) & (df_combined['median_since_last_order'] > 10), 'Customer_frequency'] = 'Regular customer'
df_combined.loc[df_combined['median_since_last_order'] <= 10, 'Customer_frequency'] = 'Frequent customer'
```
